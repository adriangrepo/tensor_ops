Numpy operations and Images

```python
import numpy as np
from PIL import Image
```

```python
#create a 2d array with 2 rows, 3 columns
x = np.array([[1,2,3], [4,5,6]])
x, x.shape
```

#### axis 0
rows

```python
for row in range(x.shape[0]):
    print(x[row, :])
```

#### axis 1
columns

```python
for col in range(x.shape[1]):
    print(x[:, col])
```

### Sum

```python
# sum entire array if axis not specified
y = x.sum()
y
```

```python
# sum columnwise: collapse along columns
y = x.sum(axis=0)
y
```

```python
# sum row-wise: collapse along rows
y = x.sum(axis=1)
y
```

### Concatenate

axis is the axis along which to join

```python
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7,8]])
```

```python
c = np.concatenate((a, b), axis=0)
c
```

```python
c = np.concatenate((a, b), axis=1)
c
```

```python
c = np.concatenate((a, b.T), axis=1)
c
```
